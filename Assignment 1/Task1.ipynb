<div style="font-size: 24px;">Data Collection</div>

```python
import urllib.request
import json
url="https://gutendex.com/books/?languages=fr%2Cfi%2Cde&page=1&popular="
response=urllib.request.urlopen(url)
raw_json=response.read().decode("utf-8")
```

Convert json data to python data type

```python
books = json.loads(raw_json)
for book in books['results']:
    print(book)
```

<div style="font-size: 18px;">Save the data as a csv file for subsequent processing</div>

```python
import csv
```

```python
selected_columns = ['id','title', 'authors','languages', 'subjects','copyright','media_type','download_count']
with open("books.csv", mode='w', newline='', encoding='utf-8') as fout:
    writer = csv.writer(fout)
    writer.writerow(selected_columns)
    for book in books['results']:
         authors = '; '.join([author['name'] for author in book['authors']])
         languages = '; '.join(book['languages'])
         subjects = '; '.join(book['subjects'])
         writer.writerow([book['id'], book['title'], authors, languages, subjects, book['copyright'], book['media_type'],book['download_count']])
```

In order to make the data statistically significant to analysis, we will repeat the collection of several more pages of data.

```python
for number in range(2, 5):
    url="https://gutendex.com/books/?languages=fr%2Cfi%2Cde&page={}&popular=".format(number)
    response=urllib.request.urlopen(url)
    raw_json=response.read().decode("utf-8")
    books = json.loads(raw_json)
    for book in books['results']:
        print(book)
    with open("books.csv", mode='a', newline='', encoding='utf-8') as fout:
        writer = csv.writer(fout)
        for book in books['results']:
             authors = '; '.join([author['name'] for author in book['authors']])
             languages = '; '.join(book['languages'])
             subjects = '; '.join(book['subjects'])
             writer.writerow([book['id'], book['title'], authors, languages, subjects, book['copyright'], book['media_type'],book['download_count']])   
```
